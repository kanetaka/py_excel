# 5章CSVを読み込む

### 問題1: c01.csvファイルをロードして上記10件を表示してください．
ヒント1：Pandas read_csv()関数を使います．

ヒント2：CSVファイルは，Shift JISです．read_csv()関数にShift JISであることを教えて上げる必要があります．

ヒント3：上記10件については，DataFrameのhead()関数を使うと簡単です．

```python
import pandas as pd

# ここに回答を書いてください．
df = pd.read_csv("c01.csv",encoding="shift_jis")
print(df[:10])
print("---")
print(df.head(10))
```

### c01.csvの内容を下から10件を表示してください．

```python
# ここに回答を書いてください．
print(df.tail(10))
```

### 下2行の注を削除して，DataFrame変数df2に代入してください．

```python
# ここに回答を書いてください．
df2 = df.drop(range(980,982))
print(df2.tail(10))
```

### 下2行を削除したデータを"deleted.csv"に保存してください．

```python
# ここに回答を書いてください．

df2.to_csv("deleted.csv", encoding="shift_jis")
```

### 下2行を削除したデータ（変数df2）で，人口（総数）を昇順で並び替えてdf_aに保存してください．

```python
df["人口（総数）"] = pd.to_numeric(df2["人口（総数）"], errors='coerce')
df_a = df2.sort_values(by="人口（総数）")
print(df_a.head(10))
```

### 下2行を削除したデータ（変数df2）で，人口（総数）を降順で並び替えてください．

```python
# ここに回答を書いてください．
df_b = df2.sort_values(by="人口（総数）", ascending=False)
print(df_b.head(10))
```

```python
df_a.to_csv("df_a.csv", encoding="shift_jis")
df_b.to_csv("df_b.csv", encoding="shift_jis")
```
